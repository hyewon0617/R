### 유용한 패키지

#### 1. doBy 패키지

* 그룹별로 데이터에 대한 기술통계량, 정렬, 표본을 추출할 떄 유용한다.
* 중요함수 
 - `summaryBy()` : 그룹별로 기술통계량을 구해주는 함수
 - `orderBy()` : 그룹별로 정렬
 - `sampleBy()` : 그룹별로 표본추출하기

```r
install.packages("doBy")
library(doBy)
```

```r
# 1. summaryBy(formula, data, FUN)
summaryBy(data=iris, formula=Sepal.Length~Species, FUN=mean)
summaryBy(data=iris, formula=Sepal.Length + Sepal.Width~Species, FUN=mean)
summaryBy(data=iris, formula=Sepal.Length~Species, FUN=sum)
summaryBy(data=iris, formula=Sepal.Length~Species, FUN=summary)
```

```r
# orderBy(fomula, data)
orderBy(~Sepal.Length, iris) # Sepal.Length기준 오름차순 정렬
orderBy(~Species + Sepal.Length, data=iris) # 품종/Sepal.Length기준 오름차순 정렬
```

```r
# 3. sampleBy(fomula, frac, repalce, systematic, data)
# frac : 추출한 표본의 비율, 기본값은 10%
# replace : 복원추출여부를 의미 TRUE=복원, FALSE=비복원
# systematic : 계층추출을 의미 TRUE=계층추출, FALSE=임의추출

# 실습: 품종별로 20%의 표본을 추출
sampleBy(~Species, frac=0.2, data=iris)
```

```r
# 통계함수 사분위수 quantile() : 0, 25, 50, 75, 100%
quantile(iris$Sepal.Length)
```

```r
# 10분위수 : 0, 10...90,100%
quantile(iris$Sepal.Length, seq(0,1,by=0.1))
```

##### 기타 유용한 함수

```r
# 1. sample(데이터셋, 샘플갯수, 중복여부) : 랜덤샘플링
# 중복여부 : replace=T or F
sample(1:10, 5)
sample(1:10, 5, replace=T) # 중복허용

# 1~10까지 10개 샘플을 뽑기(결구 무작위로 섞는 효과가 적용 됨)
sample(1:10, 10)

# iris데이터에서 150개의 샘플을 무작위로 추출
nrow(iris)
head(iris[sample(nrow(iris), nrow(iris)), ])
```

```r
# 데이터셋에서 훈련용 vs 검증용으로 분리할 경우
# iris에서 각 품종마다 30%씩 추출후 검증용으로 활용하고자 할 경우
head(sampleBy(~Species, frac=0.3, data=iris))
sam <- sampleBy(~Species, frac=0.3, data=iris)
str(sam)
```

```r
# 2. split(데이터셋, 분리조건)
split(iris, iris$Species)
```

```r
# 3. subset() : split()과 유사하지만 전체가 아닌 특정부분만 선택하는 함수
head(subset(iris, Species=="setosa"))
```

```r
# 4. merge() : 두 데이터프레임의 공통된 값을 기준으로 묶는 함수
# SQL의 join과 비슷한 코드
x <- data.frame(name=c('a','b','c'), math=c(1,2,3))
y <- data.frame(name=c('c','a','b'), english=c(4,5,6))
print(x)
print(y)

# x,y데이터셋을 name을 기준으로 연결
merge(x,y)
```

```r
# aggregate(데이터, 그룹조건, 함수) : 그룹연산함수
# iris데이터셋에서 품종별 Sepal.Width의 평균을 계산
aggregate(Sepal.Width~Species, iris, mean)
```
