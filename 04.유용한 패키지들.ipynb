##### 1. filter()

* filter()함수는 조건에 따라 행(row)을 추출한다.
* 사용방법은 첫번째인수 dataframe, 두번째인수로 조건을 지정(다른 기본함수와 동일하게 적용)
* `and 조건문` : `컴마(,)`로 구분하거나 `&연산자`를 사용
* `or 조건문`  : `|연산자`를 사용

```r
# tbl_df() : dplyr에서 사용할 데이터의 형태로 변환하는 함수
hflights_df = tbl_df(hflights)
class(hflights)
class(hflights_df)
head(hflights_df)
```

```r
# 1. and조건문 : 1월 1일 데이터추출
head(filter(hflights_df, Month==1 & DayofMonth==1))
head(filter(hflights_df, Month==1, DayofMonth==1))
head(filter(hflights, Month==1, DayofMonth==1))
```

##### 2. arrange()

```r
# transform() 
transform(hflights, gain=ArrDelay-DepDelay, gain_per_hour=gain/(AirTime/60))

# 결과
# gain과 gain_per_hour컬럼을 추가하는 데 새로 추가된 열을 같은 함수내에서 사용불가
# 생성된 gain을 gain_per_hour계산에 사용할 수 없다.
```

##### 6. group_by()

* `group_by()`함수느 지정한 열의 수준(level)별로 그룹화된 결과를 리턴

```r
# 비행편수가 20편이상, 평균비행거리가 2000마일미만인 항공사별로 평균연착시간을 계산
planes <- group_by(hflights_df, TailNum)
class(planes)
head(planes, 2)

delay <- summarise(planes,
                   count=n(), # 현재 그룹의 관측치 갯수
                   dist =mean(Distance, na.rm=TRUE) ,
                   delay=mean(ArrDelay, na.rm=TRUE))
delay <- filter(delay, count>20, dist < 2000)
head(delay,2)
```

```r
# 상기결과를 시각화
library(ggplot2)
ggplot(delay, aes(dist, delay)) +
    geom_point(aes(size=count), alpha1=1/2) +
    geom_smooth() +
    scale_size_area()
```

##### 7. chain()

* dplyr패키지의 백미인 `chain()`함수 혹은 `%>%`를 이용함으로서 각 조작을 연결해서 한번에 실행할 수 있다. 즉, 최종결과를 얻기 위해서 임시로 dataframe만들지 않아도 되는 편리한 기능이다.

###### 실습

1. Year, Month, DayOfMonth의 수준별로 그룹화
2. Year부터 DayOfMonth, ArrDelay, DepDelay열을 선택
3. 평균연착시간과 평균출발지연시간을 추출
4. 평균연착시간과 평균출발지연시간이 30분이상인 데이터를 추출

```r
# 1. chain()함수를 이용하지 않을 경우
a1 <- group_by(hflights_df, Year, Month)
a2 <- select(a1, Year:DayofMonth,ArrDelay, DepDelay)
a3 <- summarise(a2,
                arr=mean(ArrDelay, na.rm=TRUE),
                dep=mean(DepDelay, na.rm=TRUE))
a4 <- filter(a3,  arr>5 | dep>5)
head(a4,2)
```

```r
# 2. chain()함수를 이용할 경우 : %>%
hflights_df %>% 
  group_by(Year, Month, DayofMonth) %>% 
  summarise(arr=mean(ArrDelay, na.rm=TRUE), dep=mean(DepDelay, na.rm=TRUE)) %>% 
  filter(arr>30 | dep>30)
```

#### 4. stringr 

###### 기본함수(1) : 패턴을 찾아주는 함수
>* str_detect() : 정규식에 맞는 문자열의 존재여부 T/F로 리턴
* str_count() : 정규식에 맞는 문자열의 총갯수
* str_which() : 정규식에 맞는 문자열의 위치를 리턴
* str_locate() : 정규식에 맞는 문자열의 시작점과 종료지점을 리턴

###### 기본함수(2) : 문자열에서 특정부분을 리턴 함수
>* str_sub() : 문자열중 지정된 시작점과 종료지점에 해당하는 문자열을 반환
* str_subset() : 문자열중 정의된 패턴의 문자열을 반환
* str_extract() : 문자열중 해당 패턴을 vector로 반환
* str_match() : 문자열중 해당 패턴을 matrix로 반환
* str_match_all() : 첫번쨰 패턴만 찾는지 전체패턴을 찾는지의 여부

##### 1. 부분집합찾기
<img src="./images/04.유용한 패키지들_01.png" width="500" height="250">
* stringr 에서 문자열에서 부분을 반환하는 함수들은 5가지 정도가 있습니다.
* str_sub 메소드는 문자열 중 지정된 시작점과 종료지점에 해당하는 문자열을 반환하는 함수입니다.
* str_subset 메소드는 string 중에서 해당 패턴을 갖고 있는 문자열들을 반환하는 함수입니다. 다른 메소드 같은 경우는 패턴에 맞는 부분만 반환하지만 str_subset 메소드는 패턴이 있는 문자열 자체들을 반환합니다.
* str_extract 와 str_match, str_match_all 메소드는 string 중 해당 패턴을 추출하는 메소드 입니다만 str_extract 와 str_match 간에는 vertor 와 matrix 로 반환한다는 차이점이 있습니다.

또한 str_match 와 str_match_all 함수 간에는 첫 번째 패턴만 찾는지 전체 패턴을 찾는지 차이가 있습니다.
##### 2. 문자열 변형하기 
<img src="./images/04.유용한 패키지들_02.png" width="500" height="250">
* str_sub 메소드는 부분집합 찾기에서 살펴본 메소드 입니다. 그런데 str_sub 메소드를 다음과 같이 사용하면 문자열을 변형할 때 사용할 수 있습니다.

```r
# 차량별 연비기준으로 막대그프그리기
# 데이터에는 차량모델이름을 나타내는 열이 없다.
# 행이름을 확인하는 rownames()이로 열이름을 추가
mtcars$name <- rownames(mtcars)
head(mtcars)
```

```r
# 시각화
# 이전에는 ggplot()안에 속성을 넣었는데 이번에는 geom_bar()에 속성을 정의
# 이 예에서는 데이터가 하나뿐이지만 만약 서로 다른 데이터를 가져와서 그래프 하나에
# 넣고 싶을 경우 아래와 같은 식으로 사용하면 된다.
ggplot() + geom_bar(data=mtcars, aes(x=name, y=mpg), stat='identity')
```

```r
# 연비순서대로 정렬 : reorder()
# 역순으로 정렬시에는 reorder(name, -mtcars$mpg)형태로 마이너스(-)만 붙이면 된다.
ggplot(data=mtcars, aes(x=reorder(name, mtcars$mpg), y=mpg)) + geom_bar(stat='identity')
```

```r
# x축의 label의 겹침방지 : theme()적용
ggplot(data=mtcars, aes(x=reorder(name, mtcars$mpg), y=mpg)) + 
    geom_bar(stat='identity') +
    theme(axis.text.x=element_text(angle=90))
```

```r
# 그래프를 회전 : coord_flip()
ggplot(data=mtcars, aes(x=reorder(name, -mtcars$mpg), y=mpg)) + 
    geom_bar(stat='identity') +
    coord_flip()

gg <- ggplot(data=mtcars, aes(x=reorder(name, -mtcars$mpg), y=mpg)) + geom_bar(stat='identity')
```

```r
# 레이블생략
gg + coord_flip() + labs(x='', y='')
```

```r
# 그래프의 title과 특정자료를 강조
gg + coord_flip() + labs(x='', y='', title='차량별연비') +
     theme(plot.title=element_text(size=20)) +
# 특정차량모델을 다른 색으로 강조
     geom_bar(data=mtcars[mtcars$name=='Merc 450SE', ], 
              aes(x=name,y=mpg), stat='identity', fill='red') +
# 텍스트 년비를 추가
# hjust=1.25 : 그래프 끝에서 가로로 1.25떨어진 지점에 위치
     geom_text(aes(label=mpg), size=3, hjust=1.25, color='white')
```
